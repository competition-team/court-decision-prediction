````python
%load_ext autoreload
%autoreload 2
import sys
sys.path.append('../..')

from court_decision_prediction.utils import *

from sklearn.feature_extraction.text import TfidfVectorizer
from pycaret.classification import setup, compare_models, tune_model, blend_models, evaluate_model, automl, predict_model

EXP_NAME = '2_party_group_only'
````

# 1. Load Data 

````python
train_data = pd.read_csv(PATH.train)
test_data  = pd.read_csv(PATH.test)
target     = 'first_party_winner'

train_data_pp = train_data.copy()
append_party_group(train_data_pp)

test_data_pp = test_data.copy()
append_party_group(test_data_pp)

train_data_pp.head(1)
````

````python
train_data.head(2)
````

````python
from sklearn.preprocessing import OneHotEncoder

def get_vector(df, vectorizer, oh_enc, train):
    if train:
        parties = oh_enc.fit_transform(df[['first_party_grp', 'second_party_grp']])
#         facts   = vectorizer.fit_transform(df['facts'])
    else:
        parties = oh_enc.transform(df[['first_party_grp', 'second_party_grp']])
#         facts   = vectorizer.transform(df['facts'])
    rst = parties
    return np.asarray(rst)
````

````python
vectorizer = TfidfVectorizer()
oh_enc = OneHotEncoder(sparse=False)

X_train = get_vector(train_data_pp, vectorizer, oh_enc, train=True)
y_train = train_data_pp[target]

X_test  = get_vector(test_data_pp, vectorizer, oh_enc, train=False)
````

# 2. Modeling

````python
test_data_fin = test_data_pp[['first_party_grp', 'second_party_grp']]
````

````python
train_data_fin = train_data_pp[['first_party_grp', 'second_party_grp', target]]
train_data_fin
````

````python
s = setup(train_data_fin, target=target, categorical_features=['first_party_grp', 'second_party_grp'], use_gpu=True)
````

````python
base_models = compare_models(n_select=5)
````

# 3. Select Data

````python
n_zeros = len(train_data_fin[train_data_fin[target] == 0])
train_data_fin_zero = train_data_fin[train_data_fin[target] == 0]
train_data_fin_one  = train_data_fin[train_data_fin[target] == 1]
````

````python
from sklearn.preprocessing import LabelEncoder

train_data_fin_one['first_party_grp_int'] = LabelEncoder().fit_transform(train_data_fin_one['first_party_grp'])
````

````python
train_data_fin_one['first_party_grp_int']
````

````python
train_data_fin_one
````

````python
from sklearn.model_selection import train_test_split

_, train_data_fin_one = train_test_split(train_data_fin_one, test_size=n_zeros, stratify=train_data_fin_one['first_party_grp_int'])
train_data_fin_one = train_data_fin_one[train_data_fin_zero.columns]
````

````python
train_data_fin_adj = pd.concat([train_data_fin_zero, train_data_fin_one], ignore_index=True)
train_data_fin_adj
````

````python
s = setup(train_data_fin_adj, target=target, categorical_features=['first_party_grp', 'second_party_grp'], use_gpu=True)
````

````python
base_models = compare_models(n_select=5)
````

````python
# 4) Hyperparameter tuning
tuned_models = [tune_model(model, choose_better=True, return_train_score=True) for model in base_models]

# 5) Ensemble models
ensemble_model = blend_models(tuned_models, choose_better=True, return_train_score=True)  # voting(average)

# 6) Select final model
final_model = automl(return_train_score=True)  # 생성한 모델들 중 최적의 모델을 선택

# 7) Evaluate model
evaluate_model(final_model)  # 다양한 평가지표들과 시각자료들을 통해 모델을 평가

# 8) Predict
preds = predict_model(final_model, data=test_data_fin)  # 'Label' column으로 예측값을 추가
````
